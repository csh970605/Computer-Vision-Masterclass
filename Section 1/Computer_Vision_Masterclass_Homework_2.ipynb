### Haarcascade x HOG x CNN

In order to compare the results, you can use the image people3.jpg that is in the Images folder

Try to tune the parameters of each algorithm to get the best result possible. Which algorithm had the best result?

```
import dlib
import cv2
from google.colab.patches import cv2_imshow
```

#### Haarcascade

```
image = cv2.imread('people3.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
car_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
detections = car_detector.detectMultiScale(image_gray, scaleFactor=1.001, minNeighbors=2)

for x, y, w, h in detections:
  cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

cv2_imshow(image)
```

#### HOG

```
image = cv2.imread('people3.jpg')
face_detector_hog = dlib.get_frontal_face_detector()
detections = face_detector_hog(image, 5)
detections
for face in detections:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 0), 2)

cv2_imshow(image)
```

#### CNN

```
image = cv2.imread('people3.jpg')
cnn_detector = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
detections = cnn_detector(image, 4)
for face in detections:
  l, t, r, b = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 0), 2)

cv2_imshow(image)
```
