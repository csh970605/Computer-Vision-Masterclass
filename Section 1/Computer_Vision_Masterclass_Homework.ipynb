### Other objects

```
import cv2
from google.colab.patches import cv2_imshow
```

#### Cars

```
image = cv2.imread('car.jpg')
image.shape
```

```
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
car_detector = cv2.CascadeClassifier('cars.xml')
detections = car_detector.detectMultiScale(image_gray, scaleFactor=1.02, minNeighbors=6)

for x, y, w, h in detections:
  cv2.rectangle(image, (x, y), (x+w, y+h), (0,255,0), 2)

cv2_imshow(image)
```

#### Clocks

```
image = cv2.imread('clock.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
car_detector = cv2.CascadeClassifier('clocks.xml')
detections = car_detector.detectMultiScale(image_gray, scaleFactor=1.03, minNeighbors=1)

for x, y, w, h in detections:
  cv2.rectangle(image, (x, y), (x+w, y+h), (0,255,0), 2)

cv2_imshow(image)
```

#### Full body

```
image = cv2.imread('people3.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
car_detector = cv2.CascadeClassifier('fullbody.xml')
detections = car_detector.detectMultiScale(image_gray, scaleFactor=1.03, minNeighbors=1, minSize=(50, 100))

for x, y, w, h in detections:
  print(w, h)
  cv2.rectangle(image, (x, y), (x+w, y+h), (0,255,0), 2)

cv2_imshow(image)
```
